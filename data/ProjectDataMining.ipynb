```python
df = pd.read_csv("/content/cleaned_computed_insight_success_of_active_sellers.csv")

# ✅ Generate realistic classification labels with thresholds of 1500 and 8000
df['success_label'] = pd.cut(df['total_units_sold'],
                             bins=[-1, 1500, 8000, float('inf')],
                             labels=["Low", "Medium", "High"])

# Columns that need outlier treatment using log transform
columns_to_log = [
    "mean_units_sold_per_product",
    "merchant_ratings_count",
    "mean_product_prices",
    "mean_retail_prices",
    "average_discount",
    "mean_product_ratings_count",
    "urgency_text_rate"
]

# Use log1p implementation and error handling
for col in columns_to_log:
    df[col] = df[col].apply(lambda x: np.log1p(x) if pd.notnull(x) and x > -1 else np.nan)
    df[col] = df[col].fillna(df[col].median())


output_path = "/content/log_scaled_labeled_success_1500_8000.csv"
df.to_csv(output_path, index=False)
print(f"✅ Đã xử lý và lưu dữ liệu tại: {output_path}")
```

```python
df = pd.read_csv("/content/cleaned_summer-products-with-rating-and-performance_2020-08.csv")

# Create a classification label from the 'rating' column
df['rating_label_balanced'] = pd.cut(df['rating'],
                                     bins=[1.0, 3.67, 4.03, 5.0],
                                     labels=['Low', 'Medium', 'High'],
                                     include_lowest=True)

# Columns that need outlier treatment using log transform
columns_to_log = [
    "price",
    "retail_price",
    "units_sold",
    "rating_count",
    "rating_five_count",
    "rating_one_count",
    "merchant_rating_count"
]

# se log1p implementation and error handling
for col in columns_to_log:
    df[col] = np.log1p(df[col])

output_path = "/content/log_scaled_labeled_summer_products.csv"
df.to_csv(output_path, index=False)
print(f"✅ Đã xử lý bằng log và lưu dữ liệu tại: {output_path}")
```

```python
df = pd.read_csv("/content/cleaned_unique-categories.sorted-by-count.csv")

# Calculate log1p(count) to handle outlier/skew
df['log_count'] = np.log1p(df['count'])

# Split the label 'popularity_log_adjusted' from log_count
df['popularity_log_adjusted'] = pd.cut(df['log_count'],
                                       bins=[0, 1.0, 2.2, float('inf')],
                                       labels=["Rare", "Medium", "Popular"])

output_path = "/content/log_adjusted_popularity_unique_categories.csv"
df.to_csv(output_path, index=False)

print(f"✅ Đã lưu dữ liệu đã chia nhãn cân bằng tại: {output_path}")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Đọc dữ liệu
df = pd.read_csv("/content/cleaned_computed_insight_success_of_active_sellers.csv")

# Gán nhãn phân loại dựa trên total_units_sold
df['success_label'] = pd.cut(
    df['total_units_sold'],
    bins=[-1, 1500, 8000, float('inf')],
    labels=["Low", "Medium", "High"]
)

# 1. Thống kê số lượng từng nhãn
print("🔢 Distribution of 'success_label':")
print(df['success_label'].value_counts())
print("\n📊 Normalized distribution:")
print(df['success_label'].value_counts(normalize=True))

# 2. Vẽ biểu đồ cột phân phối nhãn
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='success_label', palette='pastel')
plt.title("Distribution of Success Labels")
plt.xlabel("Success Label")
plt.ylabel("Number of Records")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 3. Thống kê mô tả theo nhãn
group_stats = df.groupby('success_label')['total_units_sold'].agg(['min', 'max', 'mean', 'median', 'std'])
print("\n📈 Descriptive statistics for 'total_units_sold' by label:")
print(group_stats)

# 4. Boxplot so sánh phân phối theo nhãn
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='success_label', y='total_units_sold', palette='Set2')
plt.title("Boxplot of Total Units Sold by Success Label")
plt.xlabel("Success Label")
plt.ylabel("Total Units Sold")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
